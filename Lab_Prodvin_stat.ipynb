#### Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 

```python
MEDV = np.array(data['MEDV']).reshape(-1,1)
LSTAT = np.array(data['LSTAT']).reshape(-1,1)
model = LinearRegression()
```

1. Разделить данные на 80% тренировочных, 20% тестовых
2. Составить модель линейной регрессии
3. Посчитать ошибку модели на тестовых данных
4. Сделать точечный график данных и полученной прямой линейной регрессии

```python
# делим данные 8:2
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    MEDV,
    LSTAT,
    test_size = 0.2
)
```

```python
model.fit(X_train, y_train)
predicted_y = model.predict(X_test)
```

```python
print('Среднеквадратичная ошибка:',mean_squared_error(predicted_y, y_test))
print('Средняя абсолютная ошибка:',mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('MEDV')
plt.ylabel('LSTAT')
plt.title('Линейная регрессия от одной переменной')
x = np.linspace(-50, 200, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'yo')
plt.plot(X_test, y_test, 'b*')
plt.xlim(0,55)
plt.ylim(0,45)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

#### Смоделируйте 1000 раз сумму по 100 подбрасываний. Начертите диаграму histogram. Почему график изменился? Почему можно применить центральную предельную теорему?

```python
sum2 = []
for i in range(1000):
    array = []
    for j in range(100):
        array.append(np.random.randint(1, 6))
    sum2.append(np.sum(array))                    
plt.hist(sum2)
plt.title('Гистограмма игрального кубика')
plt.show()
```

Из-за большого количества испытаний свойство равномерности проявляется сильнее, распределение стало равномерным.

Применять центральную предельную теорему можно, благодаря закону больших чисел.
